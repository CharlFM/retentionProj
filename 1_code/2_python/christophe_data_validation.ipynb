## sales_model

```python
#assumption ID_SALES = Program_Code + ID + Travel_Type
#check if ID_SALES is unique
print("sales_model size",sales_model.shape)
print("sales_model ID_SALES unique value:",sales_model.ID_SALES.nunique())
```

```python
import re
#del new_sales_model
new_sales_model = pd.DataFrame()
new_sales_model = sales_model
for label, row in new_sales_model.iterrows():
    #print(row["ID_SALES"])    
    # CODE = label[0:2]
    #print(label,NEW_ID_SALES)
    CODE =  row["ID_SALES"]
    PCODE = row["Program_Code"]
    if not re.search('^'+PCODE, CODE):
        print ("Not Match Program_Code",CODE,PCODE)
    if not re.search(row["Travel_Type"]+'$',CODE ):
        print ("Not Match Travel_Type ",CODE,row["Travel_Type"])          
    
    NEW_TRAVEL_TYPE=CODE[-1:]
    new_sales_model.loc[label,"NEW_TRAVEL_TYPE"] = NEW_TRAVEL_TYPE
    NEW_ID_SALES = re.sub('^'+PCODE, "", CODE)    
    NEW_ID_SALES = re.sub(NEW_TRAVEL_TYPE+'$', "", NEW_ID_SALES)    
    #new_sales_model.loc[label,"NEW_ID_SALES"] = '0'+NEW_ID_SALES
    new_sales_model.loc[label,"NEW_ID_SALES"] = NEW_ID_SALES
    
    
new_sales_model.sample(100)

#difference beween last value of ID_SALES and Travel_Type
#Not Match Travel_Type  HD12110 A
#Not Match Travel_Type  HD20240 T
#Not Match Travel_Type  HD36320 T
```

## finance_model

```python
#ID_FINANCE = Special_Pay + ID
#finance_model
#finance_model.dtypes
#check if ID_SALES is unique
print("sales_model size",finance_model.shape)
print("sales_model ID_SALES unique value:",finance_model.index.nunique())
```

```python
import re
new_finance_model = pd.DataFrame()
new_finance_model = finance_model
for label, row in new_finance_model.iterrows():
    CODE = row["ID_FINANCE"]
    Special_Pay = row["Special_Pay"]
    Special_Pay = str(Special_Pay)
    #print(CODE)
    if not re.search('^'+Special_Pay, CODE):
        print ("Not Match Program_Code",CODE,Special_Pay)         
    if Special_Pay != 0:
        NEW_ID_FINANCE = re.sub('^'+Special_Pay, "", CODE)   
        #finance_model.loc[label,"NEW_ID_FINANCE"] = '0'+NEW_ID_FINANCE
        finance_model.loc[label,"NEW_ID_FINANCE"] = NEW_ID_FINANCE
   
new_finance_model.sample(100)

#Not Match Program_Code 1389 0
#Not Match Program_Code 3090 0
#Not Match Program_Code 3639 0
#Not Match Program_Code 4249 0
```
